```python
import matplotlib.pyplot as plt

# equation taken from 'manage your money like a fucking grownup in' Chapter 8                                                                
def compute_gain(P, t, r):
    F = P*(1+r/12)**(12*t)
    return F


# declare starting amount in Rands                                                                                                           
s=1000

# declare annual interest rates for different ways of storing savings                                                                        
r_etf=0.13  # long-term average growth for ETFs (inflation + ETF growth)                                                                     
r_bank=0.06  # long-term average growth of inflation in ZA (in regular savings account in ZA bank)                                           
r_bed=0.00  # cash under your bed (does not keep up with inflation so loses value every day)                                                 

etf = []
bank = []
bed = []

years = range(0, 14)
for t in years:
    etf.append(compute_gain(s, t, r_etf))
    bank.append(compute_gain(s, t, r_bank))
    bed.append(compute_gain(s, t, r_bed))

plt.plot(range(0, 14), etf, 'b', label='ETF ({0}%/year: inflation + ETF)'.format(round(r_etf*100)))
plt.plot(range(0, 14), bank, 'r', label='bank ({0}%/year: keeping up with inflation)'.format(round(r_bank*100)))
plt.plot(range(0, 14), bed, 'k-', label='under bed ({0}%/year: devaluation)'.format(round(r_bed*100)))
plt.xlabel('Years after starting with R1000')
plt.ylabel('Rands')
plt.legend()
plt.show()
```
