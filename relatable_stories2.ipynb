### define constants and compute timeseries

```python
# declare starting wealth                                                                                                       
wealth_fund=0
wealth_etf=0

# decare annnual savings in millions
annual_deposit=0.03

# declare ETF annual return                                                                        
r_etf=1.07  # 7%/year
r_fund=1.04  # 4%/year

etf = []
fund = []

years = range(0, 30)
for t in years:
    fund.append(wealth_fund)
    wealth_fund += annual_deposit
    wealth_fund = wealth_fund * r_fund
    
    etf.append(wealth_etf)
    wealth_etf += annual_deposit
    wealth_etf = wealth_etf * r_etf
```

### plot the results

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=(12, 8)) 

ax.plot(years, etf, 'b', linewidth=3, label='30K CHF/year deposit into ETF ({0}%/year)'.format(round((r_etf-1)*100)))
ax.plot(years, fund, 'r', linewidth=3, label='30K CHF/year deposit into fund ({0}%/year)'.format(round((r_fund-1)*100)))
plt.xlabel('Years after starting with 0CHF', fontsize=18)
plt.ylabel('Millions of CHF', fontsize=18)
ax.tick_params(axis='both', which='major', labelsize=14)
plt.legend(fontsize=16, loc='best')
plt.show()
```
