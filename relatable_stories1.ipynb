### define constants and compute timeseries

```python
# declare starting wealth                                                                                                       
wealth_bank=0
wealth_etf=0

# decare annnual savings in millions
annual_deposit=0.03

# declare ETF annual return                                                                        
r_etf=1.07  # 7%/year

etf = []
bank = []

years = range(0, 30)
for t in years:
    bank.append(wealth_bank)
    wealth_bank += annual_deposit
    
    etf.append(wealth_etf)
    wealth_etf += annual_deposit
    wealth_etf = wealth_etf * r_etf
```

### plot the results

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(12, 8)) 

ax.plot(years, etf, 'b', label='30K CHF/year deposit into ETF ({0}%/year)'.format(round((r_etf-1)*100)))
ax.plot(years, bank, 'r', label='30K CHF cash deposit into bank')
plt.xlabel('Years after starting with 0CHF', fontsize=18)
plt.ylabel('Millions of CHF', fontsize=18)
ax.tick_params(axis='both', which='major', labelsize=14)
plt.legend(fontsize=18)

plt.show()
```
